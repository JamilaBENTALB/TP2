```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from pandas import DataFrame
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
txt = '''Perhaps one of the most significant advances made by Arabic mathematics began at this time with the work of al-Khwarizmi, namely the 
beginnings of algebra. It is important to understand just how significant this new idea was. It was a revolutionary move away from the Greek 
concept of mathematics which was essentially geometry. Algebra was a unifying theory which allowed rational numbers, irrational numbers, 
geometrical magnitudes, etc., to all be treated as "algebraic objects". It gave mathematics a whole new development path so much broader in 
concept to that which had existed before, and provided a vehicle for future development of the subject. Another important aspect of the 
introduction of algebraic ideas was that it allowed mathematics to be applied to itself in a way which had not happened before.'''

txt_arabic="""ربما كانت أحد أهم التطورات التي قامت بها الرياضيات العربية التي بدأت في هذا الوقت بعمل الخوارزمي وهي بدايات الجبر, ومن المهم فهم كيف كانت هذه الفكرة الجديدة مهمة, فقد كانت خطوة نورية بعيدا عن المفهوم اليوناني للرياضيات التي هي في جوهرها هندسة, الجبر کان نظرية موحدة تتيح الأعداد الكسرية والأعداد اللا كسرية, والمقادير الهندسية وغيرها, أن تتعامل على أنها أجسام جبرية, وأعطت الرياضيات ككل مسارا جديدا للتطور بمفهوم أوسع بكثير من الذي كان موجودا من قبل, وقم وسيلة للتنمية في هذا الموضوع مستقبلا. وجانب آخر مهم لإدخال أفكار الجبر وهو أنه سمح بتطبيق الرياضيات على نفسها بطريقة لم تحدث من قبل"""
```

```python
sent=sent_tokenize(txt)
print("After sent_tokenize in englisg:",sent)
```

```python
vectorizer = CountVectorizer()
doc_term_matrix = vectorizer.fit_transform(sent)
```

```python
for i in range(len(sent)):
  print(sent[i])
  print(doc_term_matrix.toarray()[i])
  print('-' *100)
```

```python
print(doc_term_matrix.toarray())
print(vectorizer.get_feature_names())
```

# Term Frequency matrix english

```python
DataFrame(doc_term_matrix.toarray(),columns=vectorizer.get_feature_names())
print('*' * 200)
```

```python
sent_arabic=sent_tokenize(txt_arabic)
print("After sent_tokenize in arabic :",sent_arabic)
```

```python
vectorizer_arabic = CountVectorizer()
doc_term_matrix_arabic = vectorizer_arabic.fit_transform(sent_arabic)
```

```python
for i in range(len(sent_arabic)):
  print(sent_arabic[i])
  print(doc_term_matrix_arabic.toarray()[i])
  print('-' *100)
```

```python
print(doc_term_matrix_arabic.toarray())
print(vectorizer_arabic.get_feature_names())
```

# Term Frequency matrix arabic

```python
DataFrame(doc_term_matrix_arabic.toarray(),columns=vectorizer_arabic.get_feature_names())
```
